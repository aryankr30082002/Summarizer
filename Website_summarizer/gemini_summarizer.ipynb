```python
import requests
from bs4 import BeautifulSoup
import google.generativeai as genai
from IPython.display import Markdown, display

# Gemini API Key
genai.configure(api_key="AIzaSyDgIiFqltkWOnFZCvv7X6kgG6SJBqPIMg0")  # 🔑 Put your key here

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64)"
}

class Website:
    def __init__(self, url):
        self.url = url
        response = requests.get(url, headers=headers)
        soup = BeautifulSoup(response.content, 'html.parser')
        self.title = soup.title.string if soup.title else "No title found"

        for tag in soup(["script", "style", "img", "input"]):
            tag.decompose()

        self.text = soup.get_text(separator="\n", strip=True)

system_prompt = """
You are an AI assistant that summarizes website content.
Ignore ads, navigation menus, or irrelevant sections.
Summarize the main content only in clear, concise Markdown format.
"""

def user_prompt_for(website):
    return f"""Website Title: {website.title}

Content:
{website.text}

Summarize this content clearly in Markdown.
"""

def summarize(url):
    website = Website(url)
    prompt = system_prompt + user_prompt_for(website)

    # Use the correct model name as listed by list_models()
    model = genai.GenerativeModel('gemini-2.0-flash')  

    response = model.generate_content(prompt)
    return response.text

def display_summary(url):
    summary = summarize(url)
    display(Markdown(summary))

# Example
display_summary("https://cnn.com")
```
